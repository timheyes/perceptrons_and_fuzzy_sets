<h4>Two perceptron classifiers are trained to recognise the following classification of five patterns x with known class membership d.</h4>

<h4> The classifier is a discrete perceptron. Assign 1 to all augmented inputs. Learning constant is c=2 and the discrete perceptron learning rule is used. Initial weight vector is w.</h4>

<h3> Show that this weight vector (w) provides the correct classification of the entire training set. Plot the pattern error curve and the cycle error curve for 10 cycles (50 steps).</h3>

```python
import numpy as np
import math
import matplotlib.pyplot as plt

d = [1, -1, 1, -1, 1] #class membership
x = [np.matrix('0.4; 0.8; 0.2; 0.3; 0.2; 1'), np.matrix('1; 0.2; 0.5; 0.7; -0.5; 1'), \
     np.matrix('-0.1; 0.7; 0.3; 0.3; 0.9; 1'), np.matrix('0.2; 0.7; -0.8; 0.9; 0.3; 1'), \
     np.matrix('0.1; 0.3; 1.5; 0.9; 1.2; 1')]
c = 2 #learning constant
v = []
r = []
cr = []
pr = []
w = np.matrix('0.3350; 0.1723; -0.2102; 0.2528; -0.1133; 0.5012') #initial weight vector

for j in range(10):
    e = 0
    for i in range(5):
        #print('####iteration i is#####:', i+1)
        v = w.T * x[i] #feedforward
        #print('v is:', v)
        
        if v >= 0:
            f = 1
        elif v < 0:
            f = -1
        #print('f is', f)
        #print('d is:', d[i])
        r = d[i] - f
        #print('r is:', r)
        delta_w = c*r*x[i]
        #print('delta_w is:', delta_w)
        w = w + delta_w #weight update
        #print('w is:', w)
        
        p = 0.5*(d[i] - f)**2
        e = e + p
        pr.append(p)
        
    cr.append(e)

plt.figure()
plt.xlabel('cycle number')
plt.ylabel('error sum')
plt.plot(cr)

plt.figure(figsize=[10,5])
plt.xlabel('pattern number')
plt.ylabel('error sum')
plt.plot(pr)
plt.show()
```

<h3>In computer networks, there is an imprecise relationship between the level of bandwidth usage and the latency experienced in peer-to-peer communication.</h3>

<h3> Let X~ be a fuzzy set of use levels and Y~ be a fuzzy set of latencies with the membership function below (X, Y).</h3>

<h3>Find the Catesian product represented by the relation R~ = X ◦ Y. Now suppose we have a second fuzzy set of bandwidth usage given by x_dash.</h3>

<h3>Find S = x_dash ◦ R using Max-min composition and using max-product compostion.</h3>

```python
import numpy as np

x = np.matrix('0.2 0.5 0.8 1 0.6 0.1')
x_dash = np.matrix('0.3 0.6 0.7 0.9 1 0.5')
y = np.matrix('0.3 0.6 0.9 1 0.6 0.3')
r = np.zeros(shape=(6,6))
s = np.zeros(shape=(6,6))

for i in range(6):
    for j in range(6):
        r[i][j] = min(x.item(i), y.item(j))
        
print('Cartesian product r is:\n', r, '\n')

for i in range(6):
    s_tmp = []
    for j in range(6):
        s_tmp.append(min(x_dash.item(j), r.item(j, i)))
        #print(s_tmp)
    s[i][j] = max(s_tmp)
    
s_max_min = s[:, -1]    
    
print('max-min composition of x dash and r is:\n', s_max_min, '\n')

for i in range(6):
    s_tmp = []
    for j in range(6):
        s_tmp.append((x_dash.item(j) * r.item(j, i)))
        #print(s_tmp)
    s[i][j] = max(s_tmp)
    
s_max_product = s[:, -1]    
    
print('max product of x dash and r is:\n', s_max_product, '\n')
```
